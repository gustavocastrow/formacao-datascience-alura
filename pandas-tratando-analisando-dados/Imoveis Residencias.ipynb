## Imóveis Residenciais 

```python
import pandas as pd 
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')
```

```python
dados.head(10)
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = ['Apartamento', 'Casa', 'Casa de Condomínio', 'Casa de Vila', 'Quitinete']
```

```python
residencial
```

```python
dados.head(10)
```

```python
#Verificando no DF se os imoveis sao residenciais ou não, isso é feito
#através da funçao isin, que retora true or false
dados['Tipo'].isin(residencial)
```

```python
selecao = dados['Tipo'].isin(residencial)
```

```python
selecao 
```

```python
#Criando um novo DF só com retorno true do isin
dados_residencial = dados[selecao]
```

```python
dados_residencial
```

```python
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
#verificando o tamanho do df
dados_residencial.shape[0]
```

```python
dados.shape[0]
```

```python
dados_residencial.index = range(dados_residencial.shape[0])
```

```python
dados_residencial
```
